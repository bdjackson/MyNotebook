```python
import math
import itertools
from scipy.stats import poisson, norm
import numpy as np
%matplotlib inline
import matplotlib as mp
from matplotlib import pyplot as plt

mp.rcParams['text.usetex']=True
```

# Gaussian
First, we can explore the Gaussian distribution. We start by creating a function to draw gaussians in a uniform way.

```python
def draw_gaus_pdf(ax, x, mu, sig, col, style = '-'):
    rv = norm(mu, sig)
    ax.plot(x,
            rv.pdf(x),
            ''.join([col, style]),
            lw = 1,
            label=' : '.join(['$\mu$ = ',
                              str(mu),
                              '$\sigma$ = ',
                              '{:.2f}'.format(sig)]))
```

Now, we can draw several Gaussian pdfs with different parameters to see how these parameters affect the shape of the pdf.

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(-10, 12, 0.1)

draw_gaus_pdf(ax, x, 0, 1, 'b')
draw_gaus_pdf(ax, x, 0, 2, 'r')
draw_gaus_pdf(ax, x, -2, 1, 'g')
draw_gaus_pdf(ax, x, 2, 1, 'm')

leg = ax.legend(loc='best', frameon=False)
```

## Cumulative distribution function
We can also plot the cumulative distribution function for our Gaussian distributions above

```python
def draw_gaus_cdf(ax, x, mu, sig, col, style = '-'):
    rv = norm(mu, sig)
    ax.plot(x,
            rv.cdf(x),
            ''.join([col, style]),
            lw = 1,
            label=' : '.join(['$\mu$ = ',
                              str(mu),
                              '$\sigma$ = ',
                              '{:.2f}'.format(sig)]))
```

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(-10, 12, 0.1)

draw_gaus_cdf(ax, x, 0, 1, 'b')
draw_gaus_cdf(ax, x, 0, 2, 'r')
draw_gaus_cdf(ax, x, -2, 1, 'g')
draw_gaus_cdf(ax, x, 2, 1, 'm')

leg = ax.legend(loc='best', frameon=False)
```

# Poisson
Next, we can explore the discrete Poisson distribution

```python
def draw_pois_pmf(ax, x, mu, col):
    rv = poisson(mu)
    ax.plot(x,
            rv.pmf(x),
            ''.join([col, 'o-']),
            lw = 1,
            label=' = '.join(['$\lambda$',str(mu)]))
```

Now, we can draw several distributions with different means to see how the shape changes for different input parameters.

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 20)

draw_pois_pmf(ax, x, 1, 'b')
draw_pois_pmf(ax, x, 2, 'r')
draw_pois_pmf(ax, x, 5, 'g')
draw_pois_pmf(ax, x, 10, 'm')

leg = ax.legend(loc='best', frameon=False)
```

As $\lambda$ increases, we expect the Poisson distribution to approach a Gaussian with the same mean and variance. We can check this by potting the two distributions on top of one another for various choices of input parameters.

```python
fig, ax = plt.subplots(1, 1)

limits = (0, 20)
x = np.arange(limits[0], limits[1])
x_gaus = np.arange(limits[0], limits[1], 0.1)

def draw_pois_and_gaus(ax, x, param, col):
    draw_pois_pmf(ax,
                  x,
                  param,
                  col)
    draw_gaus_pdf(ax,
                  x_gaus,
                  param,
                  math.sqrt(param),
                  col,
                  '--')

draw_pois_and_gaus(ax, x, 1, 'b')
draw_pois_and_gaus(ax, x, 2, 'r')
draw_pois_and_gaus(ax, x, 5, 'g')
draw_pois_and_gaus(ax, x, 10, 'm')

leg = ax.legend(loc='best', frameon=False)
```

This is already looking pretty good! What happens if we increase the mean value even further.

```python
fig, ax = plt.subplots(1, 1)

limits = (0, 140)
x = np.arange(limits[0], limits[1])
x_wide = np.arange(limits[0], limits[1], 3)
x_gaus = np.arange(limits[0], limits[1], 0.5)

def draw_pois_and_gaus(ax, x, param, col):
    draw_pois_pmf(ax,
                  x,
                  param,
                  col)
    draw_gaus_pdf(ax,
                  x_gaus,
                  param,
                  math.sqrt(param),
                  col,
                  '--')

draw_pois_and_gaus(ax, x, 10, 'b')
draw_pois_and_gaus(ax, x, 20, 'r')
draw_pois_and_gaus(ax, x_wide, 50, 'g')
draw_pois_and_gaus(ax, x_wide, 100, 'm')

leg = ax.legend(loc='best', frameon=False)
```

By te time we reach values $\ge 10$, the differences in the two distributions is indeed becoming negligible.

## Cumulative density function
As with the Gaussian, we can draw the cumulative density function for the Poisson distributions

```python
def draw_pois_cdf(ax, x, mu, col):
    rv = poisson(mu)
    ax.plot(x,
            rv.cdf(x),
            ''.join([col, 'o-']),
            lw = 1,
            label=' = '.join(['$\lambda$',str(mu)]))
```

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 20)

draw_pois_cdf(ax, x, 1, 'b')
draw_pois_cdf(ax, x, 2, 'r')
draw_pois_cdf(ax, x, 5, 'g')
draw_pois_cdf(ax, x, 10, 'm')

leg = ax.legend(loc='best', frameon=False)
```
