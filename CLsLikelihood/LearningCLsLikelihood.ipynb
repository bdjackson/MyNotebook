## Background
Our likelihood is defined as:
$$
L(\mathbf{n},
\mathbf{\theta}^0|
\mu_{\mathrm{sig}},
\mathbf{b},
\mathbf{\theta}) =
P_{\mathrm{SR}}
\times
P_{\mathrm{CR}}
\times
C_{\mathrm{syst}}
$$

The $P_{\mathrm{SR}}$ and $P_{\mathrm{CR}}$ terms are modeled as the product of Poisson distributions (one term for each bin in each region), and the $C_{\mathrm{syst}}$ term is a probability density function used to constrain the nuissance parameters.

If we consider an example with a single signal region, and multiple control regions, the therms in the likelihood become:

$$
P_{\mathrm{SR}} =
P(n_\mathrm{S}|\lambda_\mathrm{S}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
$$

$$
P_{\mathrm{CR}} = 
\prod_{i\in CR}
P(n_{i}|\lambda_{i}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
$$

For the constrain term, $C_{\mathrm{syst}}$ will be modeled as a multi-dimensional Gaussian.

Putting this together gives
$$
L(\mathbf{n},
\mathbf{\theta}^0|
\mu_{\mathrm{sig}},
\mathbf{b},
\mathbf{\theta}) =
P(n_\mathrm{S}|\lambda_\mathrm{S}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
\times
\prod_{i\in CR}
P(n_{i}|\lambda_{i}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
\times
C_{\mathrm{syst}}(\mathbf{\theta}^{0},\mathbf{\theta})
$$

## Set up helper objects

### Parameter
Create an object to hold a parameter. This parameter keeps track of its initial and current values. It can also be flagged as fixed if the parameter is supposed to be constant.

```python
class parameter(object):
    def __init__(self,
                 name,
                 initial_value=0,
                 fixed=False,
                 integers=False):
        self.name = name
        self.initial_value = initial_value
        self.current_value = initial_value
        self.fixed = fixed
        self.integers = integers

    def get_value(self, initial=False):
        if initial:
            return self.initial_value
        return self.current_value

    def update_value(self, new_value):
        if self.fixed:
            print ('The value ',
                   self.name,
                   ' is fixed. I cannot update thist for you!')
            return
        self.current_value = new_value

    def get_range(self, range_frac=(0.6,1.5)):
        return (math.floor(range_frac[0]*self.current_value),
                math.ceil(range_frac[1]*self.current_value))

    def get_test_array(self, range_frac=(0.6,1.5), step_size=1):
        if self.integers:
            step_size = max(1., step_size)
            step_size = math.ceil(step_size)
        test_range = self.get_range(range_frac)
        return np.arange(test_range[0],
                         test_range[1]+step_size,
                         step_size)
    
    def get_points_to_sample(self, rel_step_size, num_points=3):
        return [self.current_value + i*rel_step_size*self.current_value
                for i in xrange(num_points)]
```

### Region
This is a class to define a region. The region holds a parameter for the measurement in the region.

```python
class region(object):
    def __init__(self, name, measurement):
        self.name = name
        self.measurement = parameter(name='_'.join(['measurement',name]),
                                     initial_value=measurement,
                                     fixed=True,
                                     integers=True)

    def get_measurement(self):
        return self.measurement.get_value(initial=False)
```

### Sample
Object to define a sample. The sample has a parameter for the sample strength and a dictionary which holds the expected number contribution in each region. The contribion in each region can be extracted using the `get_contribution` function which returns the expected value in a region accounting for the strength parameter

```python
class sample(object):
    def __init__(self, name, initial_strength=1.):
        self.name = name
        self.strength = parameter(name='_'.join(['strength', name]),
                                  initial_value=initial_strength)
        self.region_contributions = {}

    def set_contribution(self, region, contribution):
        self.region_contributions[region.name] = contribution

    def get_contribution(self, region, initial=False):
        if region.name not in self.region_contributions:
            return 0.
        return (self.strength.get_value(initial)*
                self.region_contributions[region.name])
```

## Set up helper functions

### Draw plot
Simple function that takes an axis, and a set of values to plot. The function draws the pre and post fit values to the axis.

```python
def draw_plot(ax, x, y_pre_fit=None, y_post_fit=None, **kw):
    if 'xlabel' in kw:
        ax.set_xlabel(kw['xlabel'], fontsize=14)
    if 'ylabel' in kw:
        ax.set_ylabel(kw['ylabel'], fontsize=14)

    if y_pre_fit is not None:
        ax.plot(x,
                y_pre_fit,
                'r--',
                lw=1,
                label='Pre fit')
    if y_post_fit is not None:
        ax.plot(x,
                y_post_fit,
                'b-',
                lw=1,
                label='Post fit')
```

### Test values
Function to test a list of values for a given parameter, and return the result of a test function for each test value

```python
def test_values(parameter,
                test_values,
                test_function,
                **kw):
    if (not isinstance(test_values, list) and
        not isinstance(test_values, np.ndarray)):
        test_values = [test_values]

    # getting current status of the parameter to test
    current_value = parameter.get_value()
    is_fixed = parameter.fixed

    # If this parameter is fixed, un-fix it, so we can vary it
    if is_fixed:
        parameter.fixed = False

    # update value with test value and extract the result
    test_results = []
    for tv in test_values:
        parameter.update_value(tv)
        test_results.append(test_function(**kw))

    # reset the initial state of the parameter
    parameter.update_value(current_value)
    if is_fixed:
        parameter.fixed = True

    # return the result
    return test_results
```

### Drawing functions
Simple functions to draw text boxes parameter boxes and legends

```python
text_box_props = dict(boxstyle='round',
                      facecolor='lightgray',
                      alpha=0.8)

def draw_text_box(ax, text_str):
    """
    Function to take text as textstr input parameter and draw
    it to the plot as a label
    """
    ax.text(0.05, 0.95,
            text_str,
            transform=ax.transAxes,
            fontsize=14,
            verticalalignment='top',
            bbox=text_box_props)

def draw_param_box(ax, param_list):
    """
    Function to take text as textstr input parameter and draw
    it to the plot as a label
    """
    param_text = '\n'.join(['{:s}: {:.2f}'.format(p.name.replace('_', '\_'), p.get_value())
                            for p in param_list])
    ax.text(0.95, 0.90,
            param_text,
            fontsize=14,
            verticalalignment='top',
            bbox=text_box_props)

def draw_legend(ax):
    """
    Draw a legend to the plot
    """
    leg = ax.legend(loc=1, frameon=True)
    frame = leg.get_frame()
    frame.set_color('white')
    frame.set_alpha(0.8)
```

## Likelihood function
This function defines our likelihood function. Using the tool, you can add:
- samples
- regions

Using these, members, the function computes the likelihood or NLL given the parameters and measurements.
- `fit_likelihood`: Fit the parameters to the measurements provided in the data.
- `draw_prediction`: Draw the likelihood in each region

```python
class master_likelihood(object):
    """
    This is a parent likelihood function that defines some
    generic functions to be inherrited by the more useful
    likelihoods we build up
    """
    def __init__(self):
        """
        __init__ function for MasterLikelihood
        This doesn't do anything for now
        """
        self.samples = []
        self.regions = []
        self.parameters = []

    def add_sample(self, sample):
        """
        Add a sample to the likelihood
        """
        self.samples.append(sample)
        self.parameters.append(sample.strength)

    def add_region(self, region):
        """
        Add a region to the likelihood
        """
        self.regions.append(region)
        self.parameters.append(region.measurement)

    def get_parameter(self, name):
        """
        Extract a parameter by name
        """
        for p in self.parameters:
            if p.name == name:
                return p
        return None

    def predict(self, region, initial=False):
        """
        compute the expected contribution in a region given
        the current set of parameters
        """
        return reduce(lambda a,b: a+b,
                      [s.get_contribution(region,
                                          initial=initial)
                       for s in self.samples])

    def calculate_likelihood(self, initial=False):
        """
        calculate the likelihood given the current set
        of parameters
        """
        return reduce(lambda a,b: a*b,
                      [poisson.pmf(r.measurement.current_value,
                                   self.predict(region=r,
                                                initial=initial))
                       for r in self.regions])
            
    def calculate_nll(self, initial=False):
        """
        calculate the NLL given the current set of parameters
        """
        return -np.log10(self.calculate_likelihood(initial))

    def draw_prediction(self, **kw):
        """
        Draw the likelihood in each region for values around
        the measurements
        """
        # determine the rows and colums
        num_plots = len(self.regions)
        num_rows = 1
        num_cols = num_plots
        
        # create the figure
        fig_size = (7*num_cols,5*num_rows)
        fig = plt.figure(figsize=fig_size)

        # step through the regions and draw to subplot
        for i, r in enumerate(self.regions):
            # create a subplot
            a = plt.subplot(num_rows, num_cols, i)
            
            # extract the test parameter from the region
            test_param = r.measurement
            
            # get an array of the x values
            x = r.measurement.get_test_array()
            
            # calculate the pre and post fit y values
            y_post_fit = test_values(test_param, x,
                                     self.calculate_likelihood)
            y_pre_fit = test_values(test_param, x,
                                    self.calculate_likelihood,
                                    initial=True)

            # draw the plot to the figure
            draw_plot(a, x=x, y_post_fit=y_post_fit,
                      y_pre_fit=y_pre_fit, **kw)
            
            # draw a line for the measurment
            a.axvline(r.get_measurement(),
                      color='k',
                      lw=3,
                      label='Measurement')

            # draw a legend to the screen
            text_str = ''.join(['Region: ', r.name])
            draw_text_box(a, text_str)
            draw_legend(a)
        draw_param_box(fig, self.parameters)

    def draw_parameters(self, **kw):
        """
        --
        """
        # determine the rows and colums
        num_plots = len(self.parameters)
        num_rows = (math.ceil(math.sqrt(num_plots))
                    if num_plots > 2 else 1)
        num_cols = math.ceil(num_plots/num_rows)
        
        # create the figure
        fig_size = (7*num_cols,5*num_rows)
        fig = plt.figure(figsize=fig_size)

        # step through the regions and draw to subplot
        for i, p in enumerate(self.parameters):
            param_name = p.name.replace('_', '\_')

            # create a subplot
            a = plt.subplot(num_rows, num_cols, i)
            
            # get an array of the x values
            range_frac = (kw['range_frac'] if 'range_frac' in kw
                          else (0.3,3.0))
            ax_range = p.get_range(range_frac=range_frac)
            x = p.get_test_array(range_frac=range_frac,
                                 step_size=(ax_range[1]-
                                            ax_range[0])/100.)

            # calculate the pre and post fit y values
            y_post_fit = test_values(p, x,
                                     self.calculate_nll)
            y_pre_fit = test_values(p, x,
                                    self.calculate_nll,
                                    initial=True)

            # draw the plot to the figure
            draw_plot(a, x=x, y_post_fit=y_post_fit,
                      y_pre_fit=y_pre_fit,
                      xlabel=param_name,
                      ylabel='NLL',
                      **kw)
            
            # draw a legend to the screen
            text_str = param_name
            draw_text_box(a, text_str)
            draw_legend(a)
        draw_param_box(fig, self.parameters)

    def fit_likelihood(self,
                       tolerance=0.05,
                       max_iterations=10,
                       max_retries=3):
        """
        fit the parameters of the likelihood to maximize the
        likelihood given the measurements in each region
        """
        # determine the parameters to optimize
        #   (the ones which are not fixed)
        parameters_to_optimize = [p for p in self.parameters
                                  if not p.fixed]

        initial_nll = self.calculate_nll(initial=True)
        current_nll = initial_nll
        relative_deviation = 0.1

        # local helper function to calculate the relative
        # deviation
        def calc_relative_deviation():
            return (initial_nll - current_nll)/initial_nll

        # local function to print parameter values
        def print_params():
            print '  Parameter values:'
            for p in parameters_to_optimize:
                print '    ', p.name, ': {:.4f}'.format(p.get_value())
            print '  NLL: {:.4f}'.format(current_nll)
            print
        # local function to take two lists (x,y), sort both based
        #    on the x list, and pick the `num_plots` entries with
        #    the lowest values of x
        def find_best_points(x, y, num_points):
            sorted_list = sorted(zip(x,y))[:num_points]
            sorted_x = [sl[0] for sl in sorted_list]
            sorted_y = [sl[1] for sl in sorted_list]
            return sorted_x, sorted_y

        print 'Initial values:'
        print_params()

        # iterate until the desired tolerance is reached, or
        #   the maximum number of iterations is exceded
        iteration = 0
        while (abs(relative_deviation) > tolerance and
               iteration < max_iterations):
            print 'Iteration: ', iteration
            iteration += 1
            
            # optimize each parameter independently
            # model the likelihood as a parabola for each
            # independent parameter
            for p in parameters_to_optimize:
                step_size = relative_deviation
                retry = True
                num_retries = 0
                while retry and num_retries < max_retries:
                    retry = False
                    num_retries += 1
                    
                    # compute the NLL for three points
                    values_to_test = p.get_points_to_sample(rel_step_size=step_size,
                                                            num_points=5)
                    test_nll = test_values(parameter=p,
                                           test_values=values_to_test,
                                           test_function=self.calculate_nll)
                    test_nll, values_to_test = find_best_points(test_nll,
                                                                values_to_test,
                                                                num_points = 3)
                    value_sq = [v*v for v in values_to_test]

                    # model the NLL as a parabola, and find the point where
                    # slope is zero
                    parabola_a = (((values_to_test[1]-values_to_test[2])
                                   *test_nll[0]) +
                                  ((values_to_test[2]-values_to_test[0])
                                   *test_nll[1]) +
                                  ((values_to_test[0]-values_to_test[1])
                                   *test_nll[2]))
                    parabola_b = ((value_sq[2]-value_sq[1])*test_nll[0] +
                                  (value_sq[0]-value_sq[2])*test_nll[1] +
                                  (value_sq[1]-value_sq[0])*test_nll[2])
                    new_minimum = -parabola_b/(2*parabola_a)

                    # check for failures. If the minimization broke,
                    #   reduce the step size and try again
                    if math.isnan(new_minimum):
                        print 'New minimum is NAN. Halving the step size'
                        step_size /= 2.
                        retry = True
                    
                    new_nll = test_values(parameter=p,
                                          test_values=new_minimum,
                                          test_function=self.calculate_nll)
                    if new_nll[0] > 1.3*current_nll:
                        print ''.join(['NLL using new minimum value, ',
                                       '{:.4f}'.format(new_nll[0]),
                                       ', is much larger than using old value, ',
                                       '{:.4f}'.format(current_nll), '.'])
                        print 'Halving the step size, and trying again'
                        step_size /= 2
                        retry = True

                    if math.isnan(new_nll[0]):
                        print ''.join(['NLL using new minimum value, ',
                                       '{:.4f}'.format(new_nll[0]),
                                       ', is NAN.'])
                        print 'Halving the step size, and trying again'
                        step_size /= 2
                        retry = True

    
                # Update the parameter with the value that minimized the NLL
                p.update_value(new_minimum)

            # calculate the new NLL, and determine how much
            # we moved in this iterations
            current_nll = self.calculate_nll()
            print_params()
            relative_deviation = calc_relative_deviation()
            initial_nll = current_nll
        print 'Final values:'
        print_params()
```

### Testing likelihood
Let's test that the likelihood function is doing something reasonable.
We can start with a simple example with one region and one sample.

```python
# define a region
r1 = region('r1', 67)

# define a sample and it's expected contribution in our region
s1 = sample('s1')
s1.set_contribution(r1, 50)

# construct the likelihood and add the sample and region
test_likelihood = master_likelihood()
test_likelihood.add_sample(s1)
test_likelihood.add_region(r1)

# fit the strength parameter to the measurement
test_likelihood.fit_likelihood()

# draw the prediction
test_likelihood.draw_prediction(xlabel='x',
                                ylabel='Likelihood')
test_likelihood.draw_parameters()
```

That worked quite well! Let us try again if we move away from the regeim where a Poisson is reasonably Gaussian.

```python
r1 = region('r1', 6)

s1 = sample('s1')
s1.set_contribution(r1, 3)

test_likelihood = master_likelihood()
test_likelihood.add_sample(s1)
test_likelihood.add_region(r1)

test_likelihood.fit_likelihood()

test_likelihood.draw_prediction(xlabel='x',
                                ylabel='Likelihood')
test_likelihood.draw_parameters()
```

Great. This worked again!

We have been testing with configurations where the initial strength parameter is reasonably close to the final value. What happens if we move further away, so the minimizer needs to work harder?

```python
r1 = region('r1', 67)

s1 = sample('s1', initial_strength=10)
s1.set_contribution(r1, 1)

test_likelihood = master_likelihood()
test_likelihood.add_sample(s1)
test_likelihood.add_region(r1)

test_likelihood.fit_likelihood()

test_likelihood.draw_prediction(xlabel='x',
                                ylabel='Likelihood')
test_likelihood.draw_parameters()
```

We got a couple of warning messages, but in the end, we got a reasonable fit.
