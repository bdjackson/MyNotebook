## Background
Our likelihood is defined as:
$$
L(\mathbf{n},
\mathbf{\theta}^0|
\mu_{\mathrm{sig}},
\mathbf{b},
\mathbf{\theta}) =
P_{\mathrm{SR}}
\times
P_{\mathrm{CR}}
\times
C_{\mathrm{syst}}
$$

The $P_{\mathrm{SR}}$ and $P_{\mathrm{CR}}$ terms are modeled as the product of Poisson distributions (one term for each bin in each region), and the $C_{\mathrm{syst}}$ term is a probability density function used to constrain the nuissance parameters.

If we consider an example with a single signal region, and multiple control regions, the therms in the likelihood become:

$$
P_{\mathrm{SR}} =
P(n_\mathrm{S}|\lambda_\mathrm{S}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
$$

$$
P_{\mathrm{CR}} = 
\prod_{i\in CR}
P(n_{i}|\lambda_{i}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
$$

For the constrain term, $C_{\mathrm{syst}}$ will be modeled as a multi-dimensional Gaussian.

Putting this together gives
$$
L(\mathbf{n},
\mathbf{\theta}^0|
\mu_{\mathrm{sig}},
\mathbf{b},
\mathbf{\theta}) =
P(n_\mathrm{S}|\lambda_\mathrm{S}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
\times
\prod_{i\in CR}
P(n_{i}|\lambda_{i}(\mu_\mathrm{sig}, \mathbf{b}, \mathbf{\theta}))
\times
C_{\mathrm{syst}}(\mathbf{\theta}^{0},\mathbf{\theta})
$$

### Master Likelihood
This master likelihood class will provide a template which will be used by not only the most simple likelihood, but for the somewhat more complete likelihood functions.
The features of this master likelihood class include:
- Dictionaries for storing initial and fitted parameters
- Template functions for computing the likelihood and NLL give a measurement
- Function to compute the likelihood given a different choice of parameters
- Basic helper tools used when making plots

```python
class MasterLikelihood(object):
    """
    This is a parent likelihood function that defines some
    generic functions to be inherrited by the more useful
    likelihoods we build up
    """
    def __init__(self, **kw):
        """
        __init__ function for MasterLikelihood
        This doesn't do anything for now
        """
        self.initial_params = kw
        self.fitted_params = kw

    def computeLikelihood(self, n_data, initial = False):
        """
        Function to return the likelihood of given a number
        of data events n_data
        In the parent likelihood, this just returns 1
        """
        return 1

    def computeNLL(self, n_data, initial = False):
        """
        Function to return the negative log-likelihood of
        given a number of data events n_data
        """
        return -np.log10(self.computeLikelihood(n_data, initial))

    def testParamLikelihood(self, n_data, **kw):
        """
        Compute the likelihood for a test value of b
        """
        previous_params = self.fitted_params.copy()
        for key, val in kw.items():
            if key in self.fitted_params:
                self.fitted_params[key] = val
        test_likelihood = self.computeLikelihood(n_data)
        
        # clean up and return the result
        self.fitted_params = previous_params
        return test_likelihood
    
    def testParamNLL(self, n_data, **kw):
        """
        Compute the negative log likelihood for a test value
        of b
        """
        return -np.log10(self.testParamLikelihood(n_data, **kw))

    def writeParamBoxCommon(self, ax, textstr):
        """
        Function to take text as textstr input parameter and draw
        it to the plot as a label
        """
        ax.text(0.05, 0.95,
                textstr,
                transform=ax.transAxes,
                fontsize=14,
                verticalalignment='top',
                bbox=text_box_props)
    
    def writeLegend(self, ax):
        """
        Draw a legend to the plot
        """
        leg = ax.legend(loc=1, frameon=True)
        frame = leg.get_frame()
        frame.set_color('white')
        frame.set_alpha(0.8)

```

## Simple background only likelihood
Let us start with a very simply likelihood to start building and testing some tools. We can consider a likelihood of a single control region with a single bin. With this likelihood, we can only constrain the background estimate, and not the systematics. The likelihood can be written as:
$$
L_\mathrm{first}(\mathbf{n}|\mathbf{b}) =
P(n_{CR}|\lambda_{CR}(b))
$$

For this example, we will assume the efficiency of our background process in our control region is simply $\epsilon = 0.5$. This gives us $\lambda_{CR} = \frac{1}{2}b$, giving
$$
L_\mathrm{first}(\mathbf{n}|\mathbf{b}) =
P(n|\lambda = \frac{1}{2}b)
$$


```python
class FirstSimpleBkgOnlyLikelihood(MasterLikelihood):
    def __init__(self, b = 0):
        """
        __init__ function for FirstSimpleBkgOnlyLikelihood.
        This takes an initial estimate of the b parameter describing the background normalization.
        """
        super(FirstSimpleBkgOnlyLikelihood, self).__init__(b=b)

    def computeLambda(self, initial = False):
        """
        Compute lambda from the likelihood parameters
        """
        return 0.5*(self.initial_params['b'] if initial
                    else self.fitted_params['b'])

    def computeLikelihood(self, n_data, initial = False):
        """
        Function to return the likelihood of given a number of data events n_data
        """
        return poisson.pmf(n_data, self.computeLambda(initial))

    def writeParamBox(self, ax):
        """
        write the fit parameters on the plot
        """
        textstr = ''.join(['b(post fit) = ',
                           '{:.2f}'.format(self.fitted_params['b']),
                           '\n',
                           'b(pre fit) = ',
                           '{:.2f}'.format(self.initial_params['b'])])
        self.writeParamBoxCommon(ax, textstr)
    
    def drawPlot(self, n_data, function_to_draw, **kw):
        """
        Draw the likelihood (pre and post fit) for various values
        around the measurement in the control region
        """
        fig, ax = plt.subplots(1, 1)
        if 'xlabel' in kw:
            ax.set_xlabel(kw['xlabel'], fontsize = 14)
        if 'ylabel' in kw:
            ax.set_ylabel(kw['ylabel'], fontsize = 14)

        # if no axis range provided, compute one
        ax_range = kw['ax_range'] if 'ax_range' in kw else None
        if ax_range is None:
            ax_range = (math.floor(0.60*n_data), math.ceil(1.40*n_data))
        x = np.arange(ax_range[0], ax_range[1], 1.)
        
        # Draw pre-fit results
        ax.plot(x,
                function_to_draw(x, initial = True),
                'r--',
                lw = 1,
                label = 'Pre fit')
        # Draw post-fit results
        ax.plot(x,
                function_to_draw(x),
                'b-',
                lw = 1,
                label = 'Post fit')
        # Draw a vertical line for the measured value
        ax.axvline(n_data, color = 'k', lw = 3, label = 'measured')

        # draw parameter box and legend
        self.writeParamBox(ax)
        self.writeLegend(ax)


    def drawLikelihood(self, n_data, **kw):
        """
        Draw the likelihood (pre and post fit) for various values
        around the measurement in the control region
        """
        self.drawPlot(n_data,
                      function_to_draw = self.computeLikelihood,
                      xlabel = 'x',
                      ylabel = 'Likelihood',
                      **kw)

    def drawNLL(self, n_data, **kw):
        """
        Draw the negative log-likelihood (pre and post fit) for various
        values around the measurement in the control region
        """
        self.drawPlot(n_data,
                      function_to_draw = self.computeNLL,
                      xlabel = 'x',
                      ylabel = 'NLL',
                      **kw)

    def fitLikelihood(self, n_data, tolerance = 0.05):
        """
        Fit the likelihood to a measured value from data. First, compute
        the NLL for three values of the b parameter. Then, fit a parabola
        to the three points, and find the value of b which gives a minimum
        value for the NLL
        """
        self.fitted_params = self.initial_params.copy()
        deviation_from_central_value = self.computeLambda() - n_data
        relative_deviation = abs(float(deviation_from_central_value)/
                                 self.computeLambda())
        iteration  = 0

        while relative_deviation > tolerance and iteration < 10:
            print 'iteration: ' , iteration
            print '    initial b: ', self.fitted_params['b']
            iteration += 1
            x = [abs(self.fitted_params['b'] -
                     i*float(deviation_from_central_value))
                 for i in xrange(3)]
            xsq = [val**2 for val in x]
            y = [self.testParamNLL(n_data, b = b) for b in x]
            
            # fit a parabola to the above three points.
            # We can ignore the determinant in the denominator because it
            # will cancel in the ratio
            parabola_a = ((x[1]-x[2])*y[0] +
                          (x[2]-x[0])*y[1] +
                          (x[0]-x[1])*y[2])
            parabola_b = ((xsq[2]-xsq[1])*y[0] +
                          (xsq[0]-xsq[2])*y[1] +
                          (xsq[1]-xsq[0])*y[2])

            # Update the fitted value of b with the value that minimized the NLL
            self.fitted_params['b'] = -parabola_b/(2*parabola_a)
            print '    updated b: ', self.fitted_params['b']
            print
            
            deviation_from_central_value = self.computeLambda() - n_data
            relative_deviation = abs(float(deviation_from_central_value)/
                                     self.computeLambda())

```

### Testing the simple Likelihood
Before moving on, it is important to ensure the functions constructed so far work as expected. This can be done by creating a few simple experiemnts where we observe a given number of events in our control region, and run the likelihood fit to see if we can fit the likelihood to agree with the data.

This should be a trivial exercise since we only have one observation and one unknown.

```python
# Define a measurement
measurement = 67

# Construct a likelihood function
first_test_likelihood = FirstSimpleBkgOnlyLikelihood(b = 20)

# fit the function
first_test_likelihood.fitLikelihood(measurement, tolerance = 0.001)

# draw the likelihood and NLL after the fit
first_test_likelihood.drawLikelihood(measurement)
first_test_likelihood.drawNLL(measurement)
```

The first test worked! Let us try one more example in the regiem where the Poisson is less Gaussian before moving on.

```python
# Define a measurement
measurement = 5

# Construct a likelihood function
first_test_likelihood = FirstSimpleBkgOnlyLikelihood(b = 3)

# fit the function
first_test_likelihood.fitLikelihood(measurement, tolerance = 0.001)

# draw the likelihood and NLL after the fit
first_test_likelihood.drawLikelihood(measurement, ax_range = (0, 10))
first_test_likelihood.drawNLL(measurement, ax_range = (0, 10))
```
